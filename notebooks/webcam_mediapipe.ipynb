# Webcam com OpenCV e MediaPipe ✨
Este notebook permite capturar sua webcam em tempo real e aplicar rastreamento de mãos em alta performance usando o **Google MediaPipe**.

```python
import cv2
import mediapipe as mp
import time

print(f"OpenCV version: {cv2.__version__}")
print(f"MediaPipe version: {mp.__version__}")
```

## Configuração do MediaPipe Hands
Vamos inicializar o detector de mãos.

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=2,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)
```

## Captura de Vídeo em Tempo Real
**Atenção:** Ao rodar a célula abaixo, uma janela separada do OpenCV abrirá. 
Pressione **'q'** ou **'ESC'** para fechar a janela e encerrar a captura.

```python
cap = cv2.VideoCapture(0) # 0 é geralmente a webcam padrão

if not cap.isOpened():
    print("Erro: Não foi possível acessar a webcam.")
else:
    print("Captura iniciada. Pressione 'q' na janela de vídeo para sair.")
    
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            print("Ignorando frame vazio da câmera.")
            continue

        # Inverte a imagem horizontalmente para um efeito de espelho
        # Converte de BGR (OpenCV) para RGB (MediaPipe)
        image = cv2.flip(image, 1)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Processa a imagem e encontra mãos
        results = hands.process(image_rgb)

        # Desenha as anotações das mãos na imagem
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image, 
                    hand_landmarks, 
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2), # Pontos
                    mp_drawing.DrawingSpec(color=(255, 255, 255), thickness=2) # Linhas
                )

        # Mostra o resultado na janela
        cv2.imshow('MediaPipe Hands - Pressione Q para Sair', image)

        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    print("Captura finalizada.")
```
