# Gerenciamento de Dataset 🛠️
Este notebook permite visualizar e deletar labels (gestos) específicos do seu arquivo `hand_landmarks.csv`.

```python
import pandas as pd
import os

CSV_PATH = "../storage/csv/hand_landmarks.csv"

def load_dataset():
    if os.path.exists(CSV_PATH):
        df = pd.read_csv(CSV_PATH)
        print(f"Dataset carregado. Total de amostras: {len(df)}")
        print("\nContagem atual de amostras por label:")
        print(df['label'].value_counts())
        return df
    else:
        print("Arquivo CSV não encontrado.")
        return None

df = load_dataset()
```

## Deletar Label Específica
Altere a variável `LABEL_TO_DELETE` para o nome do gesto que deseja remover completamente do dataset.

```python
LABEL_TO_DELETE = "nome_do_gesto_aqui"

if df is not None:
    if LABEL_TO_DELETE in df['label'].values:
        initial_count = len(df)
        # Mantém apenas as linhas que NÃO possuem a label informada
        df = df[df['label'] != LABEL_TO_DELETE]
        deleted_count = initial_count - len(df)
        
        # Salva o arquivo atualizado
        df.to_csv(CSV_PATH, index=False)
        
        print(f"✓ Sucesso! Deletadas {deleted_count} amostras da label '{LABEL_TO_DELETE}'.")
        print(f"Total de amostras restantes: {len(df)}")
    else:
        print(f"⚠ A label '{LABEL_TO_DELETE}' não foi encontrada no dataset.")
```
